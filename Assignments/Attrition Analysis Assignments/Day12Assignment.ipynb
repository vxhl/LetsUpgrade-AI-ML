## Step 4 : Parametric Tests 

##### 1 . ***Mann Whiteny Test*** of Attrition with the different Variables :
Format for all Hypothesis Test : 

**H0** - There is no significant difference between Attrition_Yes and Attrition_No datasets for {VARIABLE}

**H1** - There is a significant difference between Attrition_Yes and Attrition_No datasets for {VARIABLE}

```python
from scipy.stats import mannwhitneyu
stats,p = mannwhitneyu(dt_Ayes.DistanceFromHome, dt_ANo.DistanceFromHome) 
stats, p
```

As we can see the p-value `0.5317715668019634>0.05` so we accept the null hypothesis.
So H0 is accepted and H1 is rejected

```python
from scipy.stats import mannwhitneyu
stats,p = mannwhitneyu(dt_Ayes.Age, dt_ANo.Age) 
stats, p
```
