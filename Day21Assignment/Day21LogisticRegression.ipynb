***1. Logistic Regression Application on Bank_Personal_Loan_Modelling.xlsx***

### Step 1 : Importing the Different Libraries 

```python
import pandas as pd 
```

### Step 2 : Importing the Different Datasets 

```python
df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name = 1)
```

```python
df.head()
```

```python
df.shape
```

### Step 3 : Preprocessing the Dataset

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

We can see that ID and ZIP Code play no role in determining the result so we drop those columns from our set of independent features.

```python
df1 = df.drop(['ID', 'ZIP Code'], axis = 1)
df1
```

Now we have **df1** as the working dataset : 
- It has no null values
- No Duplicates
- The Categorical variables have already been encoded into 0 and 1
- Two Insignificant Columns **['ID','ZIP Code']**

### Step 4 : Defining the Dependent and the Independent Variables.

```python
# The Dependent Variable 
y = pd.DataFrame(df1['Personal Loan'], columns=['Personal Loan'])
x = df1.drop(['Personal Loan'], axis=1)
```

```python
y
```

```python
x
```

### Step 4 : Building the Logistic Regression Model

```python
import statsmodels.api as sm
```

```python
X1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y, X1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```

##### From the above summary we can conclude the following assumptions : 
The features listed below have p-values less as 0 and are therefore the most significant features :
- Income 
- Family
- Education
- CD Account
- Online
- Credit Card

Other significant features include : 
- CCAvg
- Securities


***2. Logistic Regression Application on Attrition Dataset ( general_data.csv )*** 

#### Step 1 : Importing the Dataset 

```python
data = pd.read_csv("general_data.csv")
```

```python
data.head()
```

```python
data.shape
```

#### Step 2 : Data Preprocessing

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data1 = data.dropna()
```

```python
data1.isna().sum()
```

```python
data1.columns
```

Dropping Irrelevant Features

```python
data2 = data1.drop(['EmployeeID','EmployeeCount','StandardHours','Over18'], axis = 1)
```

```python
data2.columns
```

We now encode the categorical features

```python
data2["Attrition"] = data2["Attrition"].astype('category')
data2["Attrition"] = data2["Attrition"].cat.codes
```

```python
data2["Department"] = data2["Department"].astype('category')
data2["Department"] = data2["Department"].cat.codes
```

```python
data2["Gender"] = data2["Gender"].astype('category')
data2["Gender"] = data2["Gender"].cat.codes
```

```python
data2["BusinessTravel"] = data2["BusinessTravel"].astype('category')
data2["BusinessTravel"] = data2["BusinessTravel"].cat.codes
```

```python
data2["EducationField"] = data2["EducationField"].astype('category')
data2["EducationField"] = data2["EducationField"].cat.codes
```

```python
data2["JobRole"] = data2["JobRole"].astype('category')
data2["JobRole"] = data2["JobRole"].cat.codes
```

```python
data2["MaritalStatus"] = data2["MaritalStatus"].astype('category')
data2["MaritalStatus"] = data2["MaritalStatus"].cat.codes
```

Now let us check out our encoded variables

```python
data2[['Attrition','Department','BusinessTravel','EducationField','Gender','JobRole','MaritalStatus']]
```

### Step 3 : Building the Logistic Rergression Model

```python
import statsmodels.api as sm
```

Defining the Dependent and Independent Variables

```python
y  = pd.DataFrame(data2['Attrition'],columns=['Attrition'])
```

```python
x = data2.drop(['Attrition'],axis = 1)
```

```python
x, y
```

```python
X1 = sm.add_constant(x)
```

```python
logistic = sm.Logit(y,X1)
```

```python
result = logistic.fit()
result.summary()
```
